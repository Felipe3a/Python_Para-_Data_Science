# O que é chardet?

**chardet** é uma biblioteca do Python que detecta automaticamente a **codificação de caracteres** de um texto.  
Isso é útil quando você recebe arquivos ou dados e não sabe se estão em **UTF-8**, **ISO-8859-1**, **UTF-16**, etc.

## Por que usar?
- Evita caracteres estranhos (�, Ã, ç errado).
- Descobre qual "formato de letras" o arquivo usa.
- Facilita abrir textos vindos de diferentes sistemas.

## Exemplo
```python
import chardet

data = b'\xff\xfeA\x00B\x00C\x00'
print(chardet.detect(data))

````python
# pip install chardet
import codecs
from chardet.universaldetector import UniversalDetector

caminho_do_arquivo = r"C:\Users\jabul\OneDrive\Desktop\FPScoders\Python_Para _Data_Science\Arquivo_json\V_OCORRENCIA_AMPLA.json"
detector = UniversalDetector()

with codecs.open(caminho_do_arquivo, 'rb') as arquivo: #rb esta lendo como binario em forma de bits no lugar de texto ou outra coisa 
    for linha in arquivo:
        detector.feed(linha)
        if detector.done:
            break

detector.close()
encoding_detectado = detector.result['encoding']

print(encoding_detectado)
````
