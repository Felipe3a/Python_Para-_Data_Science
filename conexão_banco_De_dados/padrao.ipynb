# Na celula a seguir veremos como conectar ao banco de dados, antes de tudo devemos instalar a biblioteca psycopg2, com o comando pip install psycopg2.

```python
import psycopg2 #primeiro importamos o psycopg2

#Em seguida passamos os parâmetros de conexão

dbname =    'postgres'             # coloque o nome do seu banco de dados
user =      'postgres'             #coloque o  usuario_do_banco 
password =  '6425'                # coloque a sua senha_do_banco      -
host =      'localhost'            #coloque o  endereço_do_banco
port =      '5432'                 #coloque a porta_de_conexão


#Agora fazemos a conexão
conn = psycopg2.connect(dbname = dbname, user=user, password=password, host=host, port=port)

cur = conn.cursor()  #Criamos um cursor para executar comandos SQL


#comandos sql transct-sql

conn.commit()  # Para confirmamos as alterações no banco de dados

# fechamos o cursor e a conexão
cur.close()
conn.close()


```
